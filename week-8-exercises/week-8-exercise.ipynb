```python
from sklearn.model_selection import train_test_split

# train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

```python
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

# K değerlerinin aralığı
Ks = 10

# KNN modelinin performansını değerlendirmek için kullanılacak metriklerin hesaplanması
#her bir k değeri için hesaplanacak olan:
mean_acc = np.zeros((Ks-1))  # Ortalama doğruluk değerlerini saklamak için bir dizi oluşturulur
std_acc = np.zeros((Ks-1))   # Doğruluk değerlerinin standart sapmasını saklamak için bir dizi oluşturulur

# K değerleri üzerinde döngü
for n in range(1, Ks): #her bir k değeri için;
    
    # KNN modeli oluşturulur ve eğitilir
    neigh = KNeighborsClassifier(n_neighbors=n).fit(X_train, y_train)
    
    # Test seti üzerinde tahmin yapılır
    y_hat = neigh.predict(X_test)
    
    # Doğruluk değeri hesaplanır ve ortalamaya eklenir
    mean_acc[n-1] = metrics.accuracy_score(y_test, y_hat)
    
    # Doğruluk değerinin standart sapması hesaplanır
    std_acc[n-1] = np.std(y_hat == y_test) / np.sqrt(y_hat.shape[0])
```

```python
print(mean_acc)
```

```python
print( "The best accuracy", mean_acc.max(), "with k=", mean_acc.argmax()+1) 
```

```python
#TP FN
#FP TN
TP= 72
TN= 34
FP=5
FN=3
```

```python
#PRECISION :Pozitif olarak tahmin edilen örnekler arasında gerçekten pozitif olanların oranı.
from sklearn.metrics import precision_score


y_test_binary = (y_test == 'M').astype(int)
y_predict_binary = (y_predict == 'M').astype(int)

# Precision değerini hesapla
precision = precision_score(y_test_binary, y_predict_binary, average='binary')
precision_v2 = TP / (TP + FP)
print("Precision:", precision)
print("Precision_V2:", precision_v2)
```

Recall or Sensitivity:

- Gerçekten pozitif olan örneklerin, pozitif olarak tahmin edilen örnekler içindeki oranı.
- Formül: TP / (TP + FN)

```python
from sklearn.metrics import recall_score

# Duyarlılık (Recall) değerini hesapla
recall = recall_score(y_test_binary, y_predict_binary, average='binary')
recall_v2 = TP / (TP + FN)
print("Recall:", recall)
print("Recall_v2:", recall_v2)
```

F1 Skoru:

- Hassasiyet ve duyarlılığın harmonik ortalaması. Hem hassasiyet hem de duyarlılık önemlidir ve F1 skoru bu ikisini birleştirir.
- Formül: 2 * (Hassasiyet * Duyarlılık) / (Hassasiyet + Duyarlılık)

```python
from sklearn.metrics import f1_score

precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1_score_v2 = 2 * (precision * recall) / (precision + recall)
f1_v1 = f1_score(y_test, y_predict, pos_label='M')
print("F1 Score:", f1_v1)
print("F1 Score_v2:", f1_score_v2)
```

Jaccard İndeksi (Jaccard Similarity Coefficient veya Intersection over Union):

- Gerçek ve tahmin edilen sınıflar arasındaki ortak örneklerin toplam örnek sayısına oranı.
- Yani; iki kümenin kesişiminin birleşimine oranını ifade eden bir benzerlik ölçüsüdür
- Formül: TP / (TP + FP + FN)

```python
from sklearn.metrics import jaccard_score

jaccard_index = jaccard_score(y_test, y_predict, pos_label='M')
jaccard_index_v2  =TP / (TP + FP + FN)
print("Jaccard İndeksi:", jaccard_index)
print("Jaccard İndeksi_v2:", jaccard_index_v2)
```

Log Kaybı (Logarithmic Loss):

- Sınıflandırma modelinin belirli bir örnek üzerindeki tahmininin gerçek değeri ile ne kadar uyuştuğunu ölçen bir ölçüdür. Daha düşük log kaybı daha iyidir.
- Formül: -1 * (1/n) * Σ(y_i * log(p_i) + (1 - y_i) * log(1 - p_i)), n: örnek sayısı, y_i: gerçek sınıf, p_i: tahmin olasılığı

```python
from sklearn.metrics import log_loss


y_prob = model.predict_proba(X_test)
# Log Kaybı hesapla
log_loss_value = log_loss(y_test, y_prob)
print("Log Kaybı:", log_loss_value)
```
