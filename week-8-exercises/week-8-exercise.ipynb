```python
from sklearn.model_selection import train_test_split

# train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

```python
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

# K deÄŸerlerinin aralÄ±ÄŸÄ±
Ks = 10

# KNN modelinin performansÄ±nÄ± deÄŸerlendirmek iÃ§in kullanÄ±lacak metriklerin hesaplanmasÄ±
#her bir k deÄŸeri iÃ§in hesaplanacak olan:
mean_acc = np.zeros((Ks-1))  # Ortalama doÄŸruluk deÄŸerlerini saklamak iÃ§in bir dizi oluÅŸturulur
std_acc = np.zeros((Ks-1))   # DoÄŸruluk deÄŸerlerinin standart sapmasÄ±nÄ± saklamak iÃ§in bir dizi oluÅŸturulur

# K deÄŸerleri Ã¼zerinde dÃ¶ngÃ¼
for n in range(1, Ks): #her bir k deÄŸeri iÃ§in;
    
    # KNN modeli oluÅŸturulur ve eÄŸitilir
    neigh = KNeighborsClassifier(n_neighbors=n).fit(X_train, y_train)
    
    # Test seti Ã¼zerinde tahmin yapÄ±lÄ±r
    y_hat = neigh.predict(X_test)
    
    # DoÄŸruluk deÄŸeri hesaplanÄ±r ve ortalamaya eklenir
    mean_acc[n-1] = metrics.accuracy_score(y_test, y_hat)
    
    # DoÄŸruluk deÄŸerinin standart sapmasÄ± hesaplanÄ±r
    std_acc[n-1] = np.std(y_hat == y_test) / np.sqrt(y_hat.shape[0])
```

```python
print(mean_acc)
```

```python
print( "The best accuracy", mean_acc.max(), "with k=", mean_acc.argmax()+1) 
```

```python
#TP FN
#FP TN
TP= 72
TN= 34
FP=5
FN=3
```

```python
#PRECISION :Pozitif olarak tahmin edilen Ã¶rnekler arasÄ±nda gerÃ§ekten pozitif olanlarÄ±n oranÄ±.
from sklearn.metrics import precision_score


y_test_binary = (y_test == 'M').astype(int)
y_predict_binary = (y_predict == 'M').astype(int)

# Precision deÄŸerini hesapla
precision = precision_score(y_test_binary, y_predict_binary, average='binary')
precision_v2 = TP / (TP + FP)
print("Precision:", precision)
print("Precision_V2:", precision_v2)
```

Recall or Sensitivity:

- GerÃ§ekten pozitif olan Ã¶rneklerin, pozitif olarak tahmin edilen Ã¶rnekler iÃ§indeki oranÄ±.
- FormÃ¼l: TP / (TP + FN)

```python
from sklearn.metrics import recall_score

# DuyarlÄ±lÄ±k (Recall) deÄŸerini hesapla
recall = recall_score(y_test_binary, y_predict_binary, average='binary')
recall_v2 = TP / (TP + FN)
print("Recall:", recall)
print("Recall_v2:", recall_v2)
```

F1 Skoru:

- Hassasiyet ve duyarlÄ±lÄ±ÄŸÄ±n harmonik ortalamasÄ±. Hem hassasiyet hem de duyarlÄ±lÄ±k Ã¶nemlidir ve F1 skoru bu ikisini birleÅŸtirir.
- FormÃ¼l: 2 * (Hassasiyet * DuyarlÄ±lÄ±k) / (Hassasiyet + DuyarlÄ±lÄ±k)

```python
from sklearn.metrics import f1_score

precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1_score_v2 = 2 * (precision * recall) / (precision + recall)
f1_v1 = f1_score(y_test, y_predict, pos_label='M')
print("F1 Score:", f1_v1)
print("F1 Score_v2:", f1_score_v2)
```

Jaccard Ä°ndeksi (Jaccard Similarity Coefficient veya Intersection over Union):

- GerÃ§ek ve tahmin edilen sÄ±nÄ±flar arasÄ±ndaki ortak Ã¶rneklerin toplam Ã¶rnek sayÄ±sÄ±na oranÄ±.
- Yani; iki kÃ¼menin kesiÅŸiminin birleÅŸimine oranÄ±nÄ± ifade eden bir benzerlik Ã¶lÃ§Ã¼sÃ¼dÃ¼r
- FormÃ¼l: TP / (TP + FP + FN)

```python
from sklearn.metrics import jaccard_score

jaccard_index = jaccard_score(y_test, y_predict, pos_label='M')
jaccard_index_v2  =TP / (TP + FP + FN)
print("Jaccard Ä°ndeksi:", jaccard_index)
print("Jaccard Ä°ndeksi_v2:", jaccard_index_v2)
```

Log KaybÄ± (Logarithmic Loss):

- SÄ±nÄ±flandÄ±rma modelinin belirli bir Ã¶rnek Ã¼zerindeki tahmininin gerÃ§ek deÄŸeri ile ne kadar uyuÅŸtuÄŸunu Ã¶lÃ§en bir Ã¶lÃ§Ã¼dÃ¼r. Daha dÃ¼ÅŸÃ¼k log kaybÄ± daha iyidir.
- FormÃ¼l: -1 * (1/n) * Î£(y_i * log(p_i) + (1 - y_i) * log(1 - p_i)), n: Ã¶rnek sayÄ±sÄ±, y_i: gerÃ§ek sÄ±nÄ±f, p_i: tahmin olasÄ±lÄ±ÄŸÄ±

```python
from sklearn.metrics import log_loss


y_prob = model.predict_proba(X_test)
# Log KaybÄ± hesapla
log_loss_value = log_loss(y_test, y_prob)
print("Log KaybÄ±:", log_loss_value)
```
