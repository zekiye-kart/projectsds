```python
def get_weather_data(city_ids):
    """
    Verilen şehir ID'leri için OpenWeatherMap API'yi çağırır ve hava durumu verilerini döndürür.
    """
    endpoint = 'http://api.openweathermap.org/data/2.5/weather'
    api_key = API_KEY
    weather_data_list = []

    for city_id in city_ids:
        params = {'id': city_id, 'appid': api_key}
        response = requests.get(endpoint, params=params)
        weather_data = response.json()
        weather_data_list.append(weather_data)

    return weather_data_list
```

```python
def create_dataframe(weather_data_list):
    """
    Verilen hava durumu verilerini içeren bir DataFrame oluşturur.
    """
    columns = ['Şehir', 'TSıcaklık (K)', 'Min Sıcaklık (K)','Max Sıcaklık (K)', 'Nem Oranı','Açıklama']
    data = []

    for weather_data in weather_data_list:
        city_name = weather_data['name']
        temperature = weather_data['main']['temp']
        temp_min= weather_data['main']['temp_min']
        temp_max= weather_data['main']['temp_max']
        humidity = weather_data['main']['humidity']
        description = weather_data['weather'][0]['description']
        data.append([city_name, temperature,temp_min,temp_max,humidity, description])

    df = pd.DataFrame(data, columns=columns)
    return df
```

```python
def get_weather_data(api_key, city_ids):
    endpoint = 'http://api.openweathermap.org/data/2.5/weather'
    weather_data = []

    for city_id in city_ids:
        params = {'id': city_id, 'appid': api_key}
        response = requests.get(endpoint, params=params)

        if response.status_code == 200:
            weather_data.append(response.json())
        else:
            print(f'Hava durumu verisi alınamadı: Şehir ID {city_id}')

    return weather_data
```

```python
def create_dataframe2(weather_data_list):
    """
    Verilen hava durumu verilerini içeren bir DataFrame oluşturur.
    """
    columns = ['Şehir', 'TSıcaklık (K)', 'Min Sıcaklık (K)', 'Max Sıcaklık (K)', 'Nem Oranı', 'Açıklama', 'Bölge']
    data = []

    for weather_data in weather_data_list:
        city_name = weather_data['name']
        temperature = weather_data['main']['temp']
        temp_min = weather_data['main']['temp_min']
        temp_max = weather_data['main']['temp_max']
        humidity = weather_data['main']['humidity']
        description = weather_data['weather'][0]['description']
        region = weather_data.get('Bölge', '')  # 'Bölge' bilgisini al, eğer yoksa boş string ata
        data.append([city_name, temperature, temp_min, temp_max, humidity, description, region])

    df = pd.DataFrame(data, columns=columns)
    return df
```

```python
def create_region_dataframe(api_key, region_city_ids):
    all_weather_data = []

    for region, city_ids in region_city_ids.items():
        region_weather_data = get_weather_data(api_key, city_ids)
        
        # Her bir veriyi eklerken hangi bölgeye ait olduğunu belirten 'region' bilgisini ekleyelim
        for data in region_weather_data:
            data['Bölge'] = region

        all_weather_data.extend(region_weather_data)

    df = create_dataframe2(all_weather_data)
    return df
```

```python
region_dataframe = create_region_dataframe(API_KEY, region_city_ids)
```
