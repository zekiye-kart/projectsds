```python
def make_frame (country:str):  #oluşturulan DataFrame'in hangi ülkenin verilerini içermesi gerektiğini belirler.
    """Take country name as parameter, return infected, recovered, deaths values as time series
    pd.DataFrame()
    
    """
    df = pd.DataFrame({
        'infected' : df_infected.loc[country],
        'recovered' : df_recovered.loc[country],
        'deaths': df_deaths.loc[country]
    }) #Bu işlem, önceki veri setlerinden belirli bir ülkeye ait ilgili sütunları seçer.
    
    df.index = pd.to_datetime(df.index, format = '%m/%d/%y' )#tarih formatına çeviriyoruz.
    #Bu satır, DataFrame'in indeksini tarih formatına çevirir. 'format' parametresi, tarihlerin hangi formatta olduğunu belirtir.
    # Bu örnekte, tarih formatı '%m/%d/%y' olarak belirtilmiştir.
    
    return df

#Bu fonksiyon, belirli bir ülkenin COVID-19 ile ilgili verilerini içeren bir zaman serisi DataFrame'i oluşturmak için kullanılabilir. 
# Örneğin, make_frame('Turkey') çağrısı, Türkiye'nin enfekte olmuş, iyileşmiş ve ölen vaka sayılarını içeren bir DataFrame döndürecektir.
    
```

### bfill (backfill) yöntemi:
eksik (NaN) değerleri doldurmak için kullanılan bir yöntemdir ve bir DataFrame veya Seri içindeki NaN değerleri, hemen sonraki değerle doldurur. İşte daha ayrıntılı bir açıklama:

- Eksik Değerlerin Belirlenmesi:
Öncelikle, bir DataFrame veya Seri üzerinde eksik değerleri (NaN) belirlemek gerekir.

- bfill Yöntemiyle Doldurma:
Eğer bir hücrede NaN değeri bulunuyorsa, bfill yöntemi, o hücreyi hemen sonraki değerle doldurur. Yani, eksik değer, hemen sonraki değerle değiştirilir.
İleri Yönlü Doldurma (forward fill) İle Karşılaştırma:

- bfill, eksik değerleri hemen sonraki değerle doldururken, forward fill (ffill) yöntemi eksik değerleri hemen önceki değerle doldurur.

```python
df_us['Rt'].replace(np.inf, np.nan).fillna(method='bfill')
#df_us['Rt'].replace(np.inf, np.nan): Bu ifade, "Rt" sütunundaki sonsuz değerleri (np.inf) NaN ile değiştirir. 
                # Yani, eğer "Rt" sütununda bir hücrede sonsuz bir değer varsa, bu değer NaN ile değiştirilir.
                
#.fillna(method='bfill'): Bu ifade, NaN değerleri backward fill yöntemiyle doldurur. 
                # Yani, NaN değeri bulunduğu hücreyi, hemen sonraki değerle doldurur. Bu, bir önceki değeri kullanarak NaN değerlerini doldurmanın bir yoludur.
                
                
#Sonuç olarak, bu kod satırı, "Rt" sütunundaki sonsuz değerleri NaN ile değiştirir ve ardından NaN değerlerini hemen sonraki değerle doldurarak DataFrame'i günceller. 
# Bu tür işlemler, veri temizleme sürecinde kullanılan yaygın tekniklerden biridir.
```

```python
df_us['Rt'].replace(np.inf, np.nan).bfill() #bir önceki kod ile aynı görevi görür.
```
