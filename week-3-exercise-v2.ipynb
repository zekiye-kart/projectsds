- Beats.Per.Minute: Şarkının dakikada kaç vuruşa sahip olduğunu ifade eden bir sayıdır. BPM (Beats Per Minute) ölçü birimidir.

- Energy: Şarkının enerji seviyesini temsil eden bir sayıdır. Daha enerjik şarkılar genellikle daha hızlı tempoya ve daha güçlü bir sese sahiptir.

- Danceability: Şarkının dans edilebilirlik seviyesini temsil eden bir sayıdır. Daha yüksek danceability değerleri, şarkının dans etmeye daha uygun olduğunu gösterebilir.

- Loudness..dB..: Şarkının ses düzeyini belirten bir sayıdır. Negatif değerler daha düşük ses düzeylerini, pozitif değerler ise daha yüksek ses düzeylerini temsil eder.

- Liveness: Şarkının canlı performans olup olmadığını belirten bir sayıdır. Daha yüksek liveness değerleri, şarkının canlı bir performansa daha yakın olduğunu gösterebilir.

- Valence.: Şarkının duygusal içeriğini ifade eden bir sayıdır. Daha yüksek valence değerleri, daha olumlu veya mutlu bir duygu durumunu temsil edebilir.

- Length.: Şarkının süresini temsil eden bir sayıdır. Genellikle dakika cinsinden ifade edilir.

- Acousticness..: Şarkının ne kadar akustik olduğunu belirten bir sayıdır. Daha yüksek acousticness değerleri, şarkının daha çok akustik enstrüman içerdiğini gösterebilir.

- Speechiness.: Şarkıdaki konuşma veya söz oranını temsil eden bir sayıdır. Daha yüksek speechiness değerleri, şarkının konuşma içerdiğini gösterebilir.

- Popularity: Şarkının popülerlik seviyesini ifade eden bir sayıdır. Daha yüksek popularity değerleri, şarkının daha popüler olduğunu gösterebilir.

 ### Korelasyon:

Korelasyon, iki değişken arasındaki ilişkinin gücünü ve yönünü ölçen bir istatistiksel ölçüdür. 
Korelasyon değeri genellikle -1 ile 1 arasında bir sayıdır.

- 1'e yaklaşan bir pozitif korelasyon, iki değişken arasında güçlü bir pozitif ilişki olduğunu gösterir.
- -1'e yaklaşan bir negatif korelasyon, iki değişken arasında güçlü bir negatif ilişki olduğunu gösterir.
- 0'a yaklaşan bir korelasyon, iki değişken arasında belirgin bir ilişki olmadığını gösterir.


- Korelasyon matrisi, bir veri setindeki tüm değişkenler arasındaki korelasyonları içeren bir matristir. 
- Isı haritası, korelasyon matrisini renklerle görselleştiren bir grafiktir.

```python
# Sadece sayısal sütunları seç
numeric_columns = df.select_dtypes(include=['number'])

# Korelasyon matrisini oluştur
correlation_matrix = numeric_columns.corr()

# Isı haritasını oluştur
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)  #sns.heatmap() ile ısı haritası,  annot=True ile korelasyon değerleri,   fmt ile sayı formatı,
plt.title("Değişkenler Arasındaki Korelasyon Matrisi")
plt.show()
```
