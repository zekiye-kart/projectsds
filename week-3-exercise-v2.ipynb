- Beats.Per.Minute: ≈ûarkƒ±nƒ±n dakikada ka√ß vuru≈üa sahip olduƒüunu ifade eden bir sayƒ±dƒ±r. BPM (Beats Per Minute) √∂l√ß√º birimidir.

- Energy: ≈ûarkƒ±nƒ±n enerji seviyesini temsil eden bir sayƒ±dƒ±r. Daha enerjik ≈üarkƒ±lar genellikle daha hƒ±zlƒ± tempoya ve daha g√º√ßl√º bir sese sahiptir.

- Danceability: ≈ûarkƒ±nƒ±n dans edilebilirlik seviyesini temsil eden bir sayƒ±dƒ±r. Daha y√ºksek danceability deƒüerleri, ≈üarkƒ±nƒ±n dans etmeye daha uygun olduƒüunu g√∂sterebilir.

- Loudness..dB..: ≈ûarkƒ±nƒ±n ses d√ºzeyini belirten bir sayƒ±dƒ±r. Negatif deƒüerler daha d√º≈ü√ºk ses d√ºzeylerini, pozitif deƒüerler ise daha y√ºksek ses d√ºzeylerini temsil eder.

- Liveness: ≈ûarkƒ±nƒ±n canlƒ± performans olup olmadƒ±ƒüƒ±nƒ± belirten bir sayƒ±dƒ±r. Daha y√ºksek liveness deƒüerleri, ≈üarkƒ±nƒ±n canlƒ± bir performansa daha yakƒ±n olduƒüunu g√∂sterebilir.

- Valence.: ≈ûarkƒ±nƒ±n duygusal i√ßeriƒüini ifade eden bir sayƒ±dƒ±r. Daha y√ºksek valence deƒüerleri, daha olumlu veya mutlu bir duygu durumunu temsil edebilir.

- Length.: ≈ûarkƒ±nƒ±n s√ºresini temsil eden bir sayƒ±dƒ±r. Genellikle dakika cinsinden ifade edilir.

- Acousticness..: ≈ûarkƒ±nƒ±n ne kadar akustik olduƒüunu belirten bir sayƒ±dƒ±r. Daha y√ºksek acousticness deƒüerleri, ≈üarkƒ±nƒ±n daha √ßok akustik enstr√ºman i√ßerdiƒüini g√∂sterebilir.

- Speechiness.: ≈ûarkƒ±daki konu≈üma veya s√∂z oranƒ±nƒ± temsil eden bir sayƒ±dƒ±r. Daha y√ºksek speechiness deƒüerleri, ≈üarkƒ±nƒ±n konu≈üma i√ßerdiƒüini g√∂sterebilir.

- Popularity: ≈ûarkƒ±nƒ±n pop√ºlerlik seviyesini ifade eden bir sayƒ±dƒ±r. Daha y√ºksek popularity deƒüerleri, ≈üarkƒ±nƒ±n daha pop√ºler olduƒüunu g√∂sterebilir.

 ### Korelasyon:

Korelasyon, iki deƒüi≈üken arasƒ±ndaki ili≈ükinin g√ºc√ºn√º ve y√∂n√ºn√º √∂l√ßen bir istatistiksel √∂l√ß√ºd√ºr. 
Korelasyon deƒüeri genellikle -1 ile 1 arasƒ±nda bir sayƒ±dƒ±r.

- 1'e yakla≈üan bir pozitif korelasyon, iki deƒüi≈üken arasƒ±nda g√º√ßl√º bir pozitif ili≈üki olduƒüunu g√∂sterir.
- -1'e yakla≈üan bir negatif korelasyon, iki deƒüi≈üken arasƒ±nda g√º√ßl√º bir negatif ili≈üki olduƒüunu g√∂sterir.
- 0'a yakla≈üan bir korelasyon, iki deƒüi≈üken arasƒ±nda belirgin bir ili≈üki olmadƒ±ƒüƒ±nƒ± g√∂sterir.


- Korelasyon matrisi, bir veri setindeki t√ºm deƒüi≈ükenler arasƒ±ndaki korelasyonlarƒ± i√ßeren bir matristir. 
- Isƒ± haritasƒ±, korelasyon matrisini renklerle g√∂rselle≈ütiren bir grafiktir.

```python
# Sadece sayƒ±sal s√ºtunlarƒ± se√ß
numeric_columns = df.select_dtypes(include=['number'])

# Korelasyon matrisini olu≈ütur
correlation_matrix = numeric_columns.corr()

# Isƒ± haritasƒ±nƒ± olu≈ütur
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)  #sns.heatmap() ile ƒ±sƒ± haritasƒ±,  annot=True ile korelasyon deƒüerleri,   fmt ile sayƒ± formatƒ±,
plt.title("Deƒüi≈ükenler Arasƒ±ndaki Korelasyon Matrisi")
plt.show()
```
