```python
from sklearn.model_selection import train_test_split
```

```python
X = customers[["Avg. Session Length", "Time on App", "Time on Website","Length of Membership"]] # bağımsız değişken
y = customers["Yearly Amount Spent"] # bağımlı değişken

X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=12)
```

```python
X_train
```

```python
X_test
```

```python
# Linear Regression modelini oluşturuyoruz ve eğitiyoruz. 

from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression() # Modeli oluşturuyoruz. 
```

```python
reg.fit(X_train, y_train) # Modeli eğitiyoruz.
```

```python
# Tahmin alıyoruz. 

y_pred = reg.predict(X_test)
```

```python
results = pd.DataFrame()
results["Ground Truth"] = y_test
results["Predictions"] = y_pred

results
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel("Y Test (ground truth)")
plt.ylabel("Y Predict (predictions)")
```

```python
# Regresyon modelini değerlendiriyoruz.
from sklearn import metrics
```

```python
mae = metrics.mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error (Ortalama Mutlak Hata): ", mae)
```

```python
mse = metrics.mean_squared_error(y_test, y_pred)
print("Mean Squared Error: ", mse)
```

```python
np.sqrt(metrics.mean_squared_error(y_test, y_pred))
```
